# 넘파이 API - ndarray 반복시키기

## 원소별로 ndarray 반복시키기

```python
# 딥러닝 강의에서 엄청 사용될 것임. 
# 이거 안쓰면 for loop 써야함. 
```

***np.repeat***

```python
# numpy.repeat(a, repeats, axis=None)
import numpy as np
x=3
rep = np.repeat(x, 2)
print(rep)
```

```python
x = np.array([1,2,3])
rep = np.repeat(x, 3)
print(rep)
```

```python
x = np.arange(4).reshape(2,2)
rep = np.repeat(x,3)  # axis를 설정하지 않아 벡터로서 다룸.   
print(rep)
```

```python
x = np.arange(4).reshape(2,2)
print(x)
print('-'*65)
rep = np.repeat(x,3, axis=0)  
print(rep)
```

```python
x = np.arange(4).reshape(2,2)
print(x)
print('-'*65)
rep = np.repeat(x,3, axis=1)  
print(rep)
```

```python
rep = np.repeat(x, repeats=[2,1], axis=0)
print(rep)          # 몇번씩 반복시킬지 결정. 
```

```python
x = np.arange(6).reshape(2,3)
repeats = np.repeat(x, repeats=[2,1,2], axis=1)
print(x)
print(repeats)
```

```python
row_vec = np.arange(4).reshape(1,-1)
print(row_vec)
print('-'*65)
rep = np.repeat(row_vec, repeats=3, axis=0)
print(rep)
```

```python
col_vec = np.arange(4).reshape(-1,1)
print(col_vec)
print(np.repeat(col_vec, repeats=3, axis=1))
```

## 전체적으로 ndarray 반복시키기

```python
# np.tile
a = np.arange(4)
print(a)
print('-'*65)
print(np.tile(a, reps=3))
```

```python
# np.tile의 경우 축을 기준으로 반복
a = np.arange(3)
print(a)
print(np.tile(a, reps=[2,2]))
```

```python
a = np.arange(6).reshape(2,3)
print(a)
print('-'*65)
print(np.tile(a, reps=[1,2]))
print('-'*65)
print(np.tile(a, reps=[2,1]))
print('-'*65)
tile = np.tile(a, reps=[2,2])
print(tile, tile.shape)
reps = np.array([2,2])
print('tile의 사이즈 : ', a.shape*reps)
```

## ndarray 반복의 응용

```python
# Application of Repetition
```

***row-wise case***

```python
# 순서
# np.sum --> np.repeat --> np.T
x = np.arange(4*3).reshape(4,3)
print(x, '\n')

x = np.sum(x, axis=0, keepdims=True)
print(x, '\n')

x = x.repeat(repeats=3, axis=0)
print(x, '\n')

x = x.T
print(x)
```

```python
x = np.arange(4*3).reshape(4,3)
print(x, '\n')
y = x.sum(0, keepdims=True).repeat(3,0).T
y
```

***Column-wise case***

```python
import numpy as np

x = np.arange(4*3).reshape(4,3)
print(x)
x = x.sum(axis=1, keepdims=True)
print(x)
x = x.repeat(repeats=2, axis=1)
print(x)
x = x.T
print(x)
```

```python
x = np.arange(4*3).reshape(4,3)
print(x, '\n')
y = x.sum(axis=1, keepdims=True).repeat(2,1).T
print(y)
```

## 좌표축 만들기 (np.meshgrid)

```python
x = np.arange(-2,3)
y = np.arange(-2,3)
print(x)
print(y, '\n')
                             # (5,)
X = x.reshape(1,-1).repeat(y.shape[0], axis=0) # row_vector
print(X)
Y = y.reshape(-1,1).repeat(x.shape[0], axis=1) # column_vector
print(Y, '\n')

Z = np.square(X) + np.square(Y)
print(Z)
```

```python
# numpy.meshgrid(*xi)
x = np.arange(-2,3) # ==> vector
y = np.arange(-2,3)

# 내가 원하는 X값들과 Y값들로 이루어진 좌표축을 구할 수 있음. 
X, Y = np.meshgrid(x,y) 
Z = np.square(X) + np.square(Y)
# print(X)
# print(Y)
print(Z)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

ax.plot_wireframe(X,Y,Z)
ax.tick_params(labelsize=20)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X, Y = np.meshgrid(x,y)
Z = np.square(X) + np.square(Y)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

ax.plot_wireframe(X,Y,Z)
ax.tick_params(labelsize=20)
```

## 스칼라의 Fully-connected Operation

***원소별로 계산을 위해***

```python
# numpy를 쓸 때 어떤 mindset을 가져야 하는지. 
# for loop을 쓰기 전에 잠깐 생각해보기. 
```

```python
x = np.arange(5)
y = np.arange(2,6)

print(x)
print(y)

for x_ in x:
    for y_ in y:
        print(x_ + y_, end=' ')
    print()
```

```python
# broadcasting이 일어나면서 덧셈이 됨.
for x_ in x:
    print(x_ + y) # 스칼라 + 벡터 
```

```python
# for loop 없이 연산하기
x = np.arange(5)
y = np.arange(2,6)
X,Y = np.meshgrid(x,y)
Z = X + Y
X,Y,Z = X.T,Y.T,Z.T

print(X)
print(Y)
print(Z)
```

***broadcasting 활용***

```python
# 차원을 일부로 확장시키기 broadcasting으로 계산해보기
x = np.arange(5)
y = np.arange(2,6)

X = x.reshape(-1,1) # (5,1) 차원을 하나 증가시킴
Y = y.reshape(1,-1) # (1,4) 차원 증가

Z = X + Y
print(Z) # (5,4)
```

```python
X.dot(Y)
```

## 벡터의 Fully-connected Operations

```python
import numpy as np
X = np.random.uniform(-5,5,(4,2))
Y = np.random.uniform(-5,5,(3,2))

for x in X:
    for y in Y:
        add = x + y
        print(add, end=' ')
    print()
```

```python
X = np.random.uniform(-5,5,(4,2))
Y = np.random.uniform(-5,5,(3,2))

"""차원을 넓히고 broadcasting"""
X = np.expand_dims(X, axis=1) # (4,1,2)
Y = np.expand_dims(Y, axis=0) # (1,3,2)
Z = X + Y
print(X.shape, Y.shape, '==>', Z.shape)
```

***Fully-connected Operations for Euclidean Distances***

```python
X = np.random.uniform(-5,5,(4,2))
Y = np.random.uniform(-5,5,(3,2))
for x in X:
    for y in Y:
        e_dist = np.sqrt(np.sum(np.square(x-y)))
        print(e_dist.round(2), end=' ')
    print()
```

```python
X = np.random.uniform(-5,5,(4,2))
Y = np.random.uniform(-5,5,(3,2))

X = np.expand_dims(X, axis=1)
Y = np.expand_dims(Y, axis=0)
                    # element-wise로 제곱 
Z = np.sqrt(np.sum(np.square(X-Y), axis=-1)) # 합할때 마지막 차원을 기준으로 더해줌.
print(Z)
print(Z.shape)
```

```python
X = np.random.uniform(-5,5,(4,3))
Y = np.random.uniform(-5,5,(3,3))
# print(X)
# print(Y)
X = np.expand_dims(X, axis=1)
Y = np.expand_dims(Y, axis=0)
                    # element-wise로 제곱 
Z = np.sqrt(np.sum(np.square(X-Y), axis=-1)) # 합할때 마지막 차원을 기준으로 더해줌.
print(Z)
print(Z.shape)
```
