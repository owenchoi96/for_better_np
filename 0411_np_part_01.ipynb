## 넘파이 메모리 최적화

```python
"""memory optimization"""
```

```python
# 넘파이가 어떻게 최적화를 하는지? 
# 쓸데없는 복사는 최대한 줄여서 메모리를 사용해보자. 
import numpy as np
a = np.arange(5)
b = a.view() # 같은 메모리에 배치

b[0] = 100

print(a)
print(b)
```

```python
a = np.arange(5)
b = a[0:3]
b[...] = 10

print(a)
print(b)
```

```python
# copy를 사용하면 아예 다른 메모리를 갖게 됨
# 때문에 원본에 영향을 미치지 않음. 
# view는 거의 사용하지 않음.

a = np.arange(5)
b = a.copy()
c = a.view()
d = a[0:3]

print(b.base is a) # copy만 다른 경우. 다른 메모리를 이용함. 
print(c.base is a)
print(d.base is a)
```

```python
# ndarray.reshape(), ndarray.ravel() => view가 기본
# ndarray.resize(), ndarray.flatten() => copy가 기본
```

## 브로드 캐스팅

```python
# shape이 일정하지 않을 때 연산이 가능
# 하지만 연산이 일정하지 않을 때도 연산을 하기 때문에 bug가 쉽게 나타남. 
    # 결과가 내가 원하는 모습이 안나올때가 많이 나타나기도 함. 
    
# 차원이 같을 때나 다를 때 연산이 다르게 나타남. 
```

### ndim이 1일 때의 브로드 캐스팅

```python
# 나머지 tensor에 대해서 맞춰줌
# (3,3) + (1,3) ==> (3,3) + (3,3)
import numpy as np
a = np.arange(9).reshape(3,3)
b = 10*np.arange(3).reshape(-1,3)
c = a + b

print("a: {}/{}\n{}".format(a.ndim, a.shape, a))
print("b: {}/{}\n{}\n".format(b.ndim, b.shape, b))
print("a+b: {}/{}\n{}".format(c.ndim, c.shape, c)) # 맞춰짐. 
```

```python
import numpy as np
a = np.arange(9).reshape(3,3)
b = 10*np.arange(3).reshape(-1,1)
# b = 10*np.arange(3).reshape(3,-1)
c = a + b

print("a: {}/{}\n{}".format(a.ndim, a.shape, a))
print("b: {}/{}\n{}\n".format(b.ndim, b.shape, b))
print("a+b: {}/{}\n{}".format(c.ndim, c.shape, c)) # 복사에 의하여 맞춰짐. 
```

```python
# when ndims are equal (matrices)
# (3,1) + (1,3)
a = np.arange(3).reshape(3,-1)
b = 10*np.arange(3).reshape(-1,3)
a+b
```

### ndim이 2일때 broadcasting

```python
# (2,3,3) + (2,1,3)
# ==> (2,3,3) + (2,3,3)

a = np.arange(18).reshape(2,3,3)
b = 10*np.arange(6).reshape(2,1,3)
a + b
```

```python
# (2,3,3) + (2,3,1)
# ==> (2,3,3) + (2,3,3)

a = np.arange(18).reshape(2,3,3)
b = np.arange(6).reshape(2,3,1)
a + b
```

## indexing and slicing

```python
import numpy as np
a = np.arange(10)
print(a, '\n')

print('a[0]', a[0])
print('a[2]', a[2], '\n')

print('a[-1]', a[-1])
print('a[-2]', a[-2], '\n')
```

```python
a = np.arange(10)
print(a, '\n')

print("a[0:3] ==> " , a[0:3])
print("a[3:7] ==> " , a[3:7])
print("a[5:-1] ==> " , a[5:-1])
```

```python
a = np.arange(10)
print(a, '\n')

print('a[2:] --> ', a[2:])
print('a[-2:] --> ', a[-2:])
print('a[5:] --> ', a[5:])
print('a[:-3] --> ', a[:-3])
print('a[:] --> ', a[:])
```

```python
a = np.arange(10)
print(a, '\n')

print('a[2:7:2] ==> ', a[2:7:2])
print('a[::3] ==>', a[::3], '\n')

# 반대방향
print('a[7:2:-1] ==> ', a[7:2:-1]) # -1이면 endpoint에서 +1을 한 지점까지 나타냄.
print('a[::-1] ==> ', a[::-1]) # 반대 방향으로 sort가능함. 
print('a[8:3:-1] ==> ', a[8:3:-2]) # -1이면 endpoint에서 +1을 한 지점까지 나타냄.
print('a[::-3] ==> ', a[::-3])
```

```python
a = np.arange(10)
indices = np.array([0,3,6,-1])
print(a, '\n')

print(a[[0,3,6,-1]])
print(a[indices])
print(a[a%3==0]) # 나머지가 0인 부분. (수식을 넣어줘서 인덱싱 해주기)
```

```python
a = np.arange(1,11)
print(a, '\n')
indices = np.arange(5)
a[indices] = 0 # indices에 해당하는 부분 0으로 바꾸기
print('changed a')
print(a) 

"""
numpy는 최대한 for문을 없애기 위해 나온 라이브러리이기 때문에
최대한 잘 활용하여 프로그래밍을 효율적으로 하기! 
"""
```

```python
a = np.arange(1,11)
print(a, '\n')

a[::2] = 200
print('change a')
print(a, '\n')

a[5:-1:3] = 300
print('changed again')
print(a)
```
